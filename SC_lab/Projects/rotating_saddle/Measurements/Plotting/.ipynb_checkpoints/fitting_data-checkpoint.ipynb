```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
    
sigma =0.02
mean=1.153

def normal(x_val):
    return  1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x_val-mean)**2/(2*sigma**2))

def lorentz(x_val):
    x_lor=(x_val-mean)/(sigma/2)
    return 1/(x_lor**2+1)
                             

def model(xval):
    model_int, error = quad(lorentz, 0, xval)
    model_max, error = quad(lorentz, 0, 2)   #normalize to model at 1.5 rps, where data has max.
    return model_int/model_max


x=np.linspace(0,3,100)
y=np.zeros(x.size)

for i in range(x.size):
    y[i]=model(x[i])
    
#y=y/max(y)
#liquid nitrogen data from 2/14 or thereabouts.
rps_1 = np.array([0.950, 1.030, 1.155, 1.201, 1.292, 1.379, 1.657, 1.858, 2.060, 1.107, 1.520, 2.531, 3.01])
data_1 = np.array([ 1.52      ,  1.64      ,  9.7       , 15.13333333, 15.8       ,
       13.26666667,  6.51428571,  3.18888889,  2.5375    ,  2.85555556,
        8.34285714,  2.4       ,  2.04285714])
                  
#Steel ball 19 mm from 2/24/21
rps_2 = np.array([0.607, 0.7, 0.845, 0.983, 1.161, 1.191, 1.22, 1.263, 1.31, 1.406, 1.516, 1.619])
data_2 = np.array([ 1.12,  1.05,  1.25,  1.43,  2.35,  3.61, 13.69, 24.46, 26.21, 33.42, 40.36, 35.78])

#Steel ball 6.3 mm from 2/24/21
rps_3 = np.array([0.632, 0.735, 0.843, 0.972, 1.073, 1.162, 1.192, 1.225, 1.266,
        1.325, 1.476, 1.611])
data_3 = np.array([1.18, 1.19, 1.47, 1.75, 1.59, 3.12, 5.05, 7.08, 7.06, 7.58, 7.64,
        7.63])

#Steel ball 28.5 mm from 2/24/21
rps_4 = np.array([0.61 , 0.705, 0.832, 0.974, 1.094, 1.165, 1.231, 1.28 , 1.315,
        1.403, 1.527, 1.618])
data_4 = np.array([ 0.94,  1.2 ,  1.02,  1.1 ,  1.58,  1.74,  3.86, 20.6 , 25.2 ,
        34.2 , 35.4 , 39.6 ])

rps = rps_1
data = data_1

data = data - min(data)
data = data/max(data)

squ_err=np.zeros(101)
mean_array = np.linspace(1.10, 1.20, 101)

for j in range(mean_array.size):
    mean=mean_array[j]
    for i in range(6):
        squ_err[j] = squ_err[j] + (model(rps[i])-data[i])**2
    
    
print(min(squ_err))

fig = plt.figure(figsize=(12, 8))
plt.title('LN2 f0=1.153')
plt.xlabel('RPS(1/s)')
plt.ylabel('normalized lifetime')
#     plt.errorbar(rps_1, mean_times_1,   yerr=error_bars_1, fmt='or')
plt.plot(x,y)
plt.plot(rps, data, 'ro')
#plt.plot(mean_array,squ_err, 'o')
plt.show


fig = plt.figure(figsize=(12, 8))
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(mean_array,squ_err, 'o')
plt.show
```
